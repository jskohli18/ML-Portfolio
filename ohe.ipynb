# Method 1: Using Pandas get_dummies
Since the model can't read text, converting the town names into separate dummy columns with 1s and 0s.

```python
dummies = pd.get_dummies(df.town).astype(int)
merged = pd.concat([df, dummies], axis = 'columns')
merged
```

```python
merged2 = merged.drop(['town'], axis='columns')
merged2
```

```python
final = merged2.drop(['west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
x = final.drop(['price'], axis='columns')
x
```

```python
y = final['price']
y
```

```python
model.fit(x,y)
```

```python
model.predict([[3000, 0, 0]])
```

# Method 2: Using Scikit-Learn (The Industry Way)
Using `LabelEncoder` to turn the town names into numbers (0, 1, 2).

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
x = dfle[['town','area']].values
x
```

```python
y = dfle['price'].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])],
                    remainder='passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:]
```

```python
x
```

```python
model.fit(x,y)
```

```python
model.predict([[1,0,2800]])
```
